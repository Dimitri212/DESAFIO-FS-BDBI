#Plotagem de Gráficos

-Gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department)

```python
df2= df.groupby('Department')[['EmployeeCount']].count()
df2
```

```python
df2.plot(kind = 'bar')
```

-Gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education)

```python
df3= df.groupby('Education')[['EmployeeCount']].count()
df3
```

```python
df3.plot(kind='line', color = 'c', marker = '.')
```

-Gráfico de torta/pizza (pie) que exiba a quantidade de funcionários que tem ou não algum atrito na empresa (coluna Attrition)

```python
df4= df.groupby('Attrition')[['EmployeeCount']].count()
df4
```

```python
df4.plot(kind = 'pie', subplots = True)
```
